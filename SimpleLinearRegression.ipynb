#Let's import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#Load Dataset

```python
dataset=pd.read_csv('Salary_Data.csv')
```

Split Feature and Target variable & convert to numpy from pandas

```python
X=dataset.iloc[:,:-1].values #YearsExperience
y=dataset.iloc[:,-1].values #Salary
#here we have only 2 columns
```

We don't need feature scaling as we have only one feature

#Split the dataset in using train_test_split from sklearn.model_selection

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

#Train the model using training data

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

#predict the result using trained model for X_test

```python
y_pred =regressor.predict(X_test)
```

#Plotting the traing and test datasets

```python
plt.scatter(X_train,y_train,color='red',label="Training_Data")
plt.plot(X_train,regressor.predict(X_train),color='blue',label="Regression_Line")
plt.title("Salary vs Experience (Training set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
sns.regplot(x=X_train,y=y_train,color='red') #regression line shortcut
```

#Plot Testset

```python
plt.scatter(X_test,y_test,color='red',label="TestData")
#plot regression line
plt.plot(X_test,regressor.predict(X_test),color='blue',label="Regression_Line")
plt.title("Salary vs Experience (Test set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
